## Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Scatter Plot can be done using two methods
# Scatter plot using lineplot method
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y, 'o', color = 'black')  # use 'o' for scatter plot
plt.show()
```

```python
# Scatter Plot using scatter function
plt.scatter(x,y)
plt.show()
```

```python
# styling
plt.plot(x,y, '-o', color = 'black')
plt.show()
```

```python
plt.plot(x,y, '-p', color = 'black')  # p represents hexagon shape
plt.show()
```

```python
# This plot can be used in showing diseases
x = np.random.randn(100)   # randn is a standard normal distribution
y = np.random.randn(100)    # 100 random array
colors = np.random.rand(100)      # build array of color

sizes = 1000 * np.random.rand(100)   # multiply each value by 1000

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)

plt.colorbar()

plt.show()
```

```python
#Histogram
data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, bins=30, alpha=0.5, histtype='stepfilled', color='steelblue')
plt.show()
```

```python
# This histogram is useful in showing countries wise covid cases
# This distribution can be done using histogram or bar plot
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

dict_extra = dict(histtype='stepfilled', alpha=0.3, bins=30)

plt.hist(x1, **dict_extra) # single * represents list and pass as a list while double ** represents dictionary and pass as a dictionary
plt.hist(x2, **dict_extra)
plt.hist(x3, **dict_extra)

plt.show()
```

```python
# Line graph is useful in showing heartbeat rate
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), '-b', label = "Sine Graph")
plt.plot(x, np.cos(x), '--r', label = "Cosine Graph")

plt.legend(loc='best', ncol=2, fancybox=True, borderpad=1, shadow=True)
plt.show()
```

```python
#using input
x = np.linspace(0, 10, 1000)
inp = input("Enter function Sin or Cos\n")
if inp == "Sin":
    plt.plot(x, np.sin(x), '-b', label = "Sine Graph")
else:
    plt.plot(x, np.cos(x), '--r', label = "Cosine Graph")

plt.legend()
```

```python
# 2D plot histogram
# This graph is useful in making x-rays
x, y = np.random.multivariate_normal([0,0],[[1,1],[1,2]], 10000).T    # Google
plt.hist2d(x, y, bins=20, cmap='Blues')

plt.show()
```
